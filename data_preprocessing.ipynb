```python
import pandas as pd
```

```python
df = pd.read_csv("gpa_dataset.csv")
```

```python
df = df[df["Subject"] == "CS"]
```

```python
df = df[(df["Term"] == "Fall") | (df["Term"] == "Spring")]
```

```python
df = df.reset_index()
```

```python
df = df.drop(["index"], axis = 1)
```

```python
df = df.drop(["Primary Instructor"], axis = 1)
```

```python
df = df.drop(["W"], axis = 1)
```

```python
df = df.drop(["Year", "Term", "Subject"], axis = 1)
```

```python
df
```

```python
count = df["Number"].count()
hash_map = {}
# df['Mix'] = df[['YearTerm', 'Number']].apply(lambda x: ''.join(x), axis=1)
df['Mix'] = df['Number'].astype(str) + df['YearTerm']
df
```

```python
for i in range(count):
    value = df["Primary Instructor"][i]
    key = df["Mix"][i]
    if key not in hash_map:
        hash_map[key] = value
hash_map
```

```python
df2 = df.groupby(['YearTerm','Number', 'Course Title', 'Mix'], as_index=False)['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D', 'D-', 'F'].sum()
```

```python
df2
```

```python
count = df2['Number'].count()
df2["Instructor"] = df2["Course Title"]
for i in range(count):
    key = df2['Mix'][i]
    df2['Instructor'][i] = hash_map[key]
```

```python
df2
```

```python
df2 = df2[df2["Number"] < 500]
```

```python
df2 = df2.sort_values(by=["Number", "YearTerm"])
```

```python
df2["Year"] = df2["YearTerm"].str.slice(stop=4)
```

```python
df2
```

```python
course_numbers = df2["Number"].unique()
```

```python
course_numbers
```

```python
drop_courses = [100, 199, 231, 232, 398, 410, 414, 419, 420, 422, 423, 424, 426, 427, 428, 429, 431, 433, 445, 446, 447, 450, 457, 460, 463, 466, 467, 473, 477, 484, 492, 493, 494, 498]
for course in drop_courses:
    df2 = df2[df2["Number"] != course]
```

```python
df2[df2["Number"] == 498]
```

```python
df2 = df2.sort_values(by = "Number")
df2 = df2.reset_index()
```

```python
df2["Total"] = df2["A+"] + df2["A"] + df2["A-"] + df2["B+"] + df2["B"] + df2["B-"] + df2["C+"] + df2["C"] + df2["C-"] + df2["D+"] + df2["D"] + df2["D-"] + df2["F"]
```

```python
df2["Average A"] = df2["A+"] + df2["A"]
```

```python
df2 = df2.drop(["index"], axis = 1)
```

```python
df2
```

```python
df2["GPA"] = 4 * (df2["A+"] + df2["A"]) + 3.67 * df2["A-"] + 3.33 * df2["B+"] + 3 * df2["B"] + 2.67 * df2["B-"] + 2.33 * df2["C+"] + 2 * df2["C"] + 1.67 * df2["C-"] + 1.33 * df2["D+"] + 1 * df2["D"] + 0.67 * df2["D-"]
```

```python
df2["GPA"] = df2["GPA"] / (df2["Total"])
```

```python
df2["Average A"] = df2["Average A"]/df2["Total"] * 100
```

```python
df2[df2["Number"] == 374].sort_values(by="Year")
```

```python
df2["Term"] = df2["YearTerm"].str.slice(start=5)
df2 = df2.drop(["Mix"], axis = 1)
```

```python
df2
```

```python
df2.to_csv("cs_data_instructor.csv", index = False)
```
